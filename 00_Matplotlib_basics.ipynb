# **Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load sample datasets
tips = sns.load_dataset("tips")
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

## **1. Basic Plotting**

### Line plot

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

### Scatter plot

```python
plt.scatter(iris_df["sepal length (cm)"], iris_df["sepal width (cm)"])
plt.title("Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

### Bar plot

```python
tips.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

### Histogram

```python
plt.hist(tips["total_bill"], bins=20)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### Box plot

```python
plt.boxplot([tips[tips["sex"] == "Male"]["total_bill"], tips[tips["sex"] == "Female"]["total_bill"]])
plt.title("Boxplot of Total Bill by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
plt.xticks([1, 2], ["Male", "Female"])
plt.show()
```

### Pie chart

```python
sizes = tips["day"].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Days")
plt.show()
```

### Error bar

```python
x = np.arange(0, 10, 1)
y = np.exp(x)

yerr = y * 0.1

plt.errorbar(x, y, yerr=yerr, fmt='o')

plt.title("Error Bar Example")
plt.xlabel("X")
plt.ylabel("Exp(X)")
plt.show()
```

## **2. Customization**

### Titles and labels

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

### Legends

```python
plt.plot(iris_df["sepal length (cm)"], label="Sepal Length")
plt.plot(iris_df["sepal width (cm)"], label="Sepal Width")
plt.title("Sepal Length and Width")
plt.xlabel("Index")
plt.ylabel("Size (cm)")
plt.legend()
plt.show()
```

### Axes

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.xlim(0, 150)
plt.ylim(4, 8)
plt.show()
```

### Grids

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()
```

### Subplots

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(iris_df["sepal length (cm)"])
axs[0, 0].set_title("Sepal Length")

axs[0, 1].plot(iris_df["sepal width (cm)"], 'tab:orange')
axs[0, 1].set_title("Sepal Width")

axs[1, 0].plot(iris_df["petal length (cm)"], 'tab:green')
axs[1, 0].set_title("Petal Length")

axs[1, 1].plot(iris_df["petal width (cm)"], 'tab:red')
axs[1, 1].set_title("Petal Width")

for ax in axs.flat:
    ax.set(xlabel='Index', ylabel='Size (cm)')

for ax in axs.flat:
    ax.label_outer()

plt.show()
```

### Styles

```python
plt.style.use('ggplot')
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()

plt.style.use('default')
```

## **3. Advanced Plotting**

### Logarithmic plot

```python
x = np.linspace(0.1, 10, 100)
y = np.exp(x)
plt.semilogy(x, y)
plt.title("Semilog Plot")
plt.xlabel("X")
plt.ylabel("Exp(X)")
plt.show()
```

### Polar plot

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(theta))
plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

### 3D plotting

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris_df["sepal length (cm)"], iris_df["sepal width (cm)"], iris_df["petal length (cm)"])
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")
ax.set_zlabel("Petal Length (cm)")
plt.title("3D Scatter Plot")
plt.show()
```

### Date plotting

```python
dates = pd.date_range('2023-01-01', periods=100)
data = np.random.randn(100).cumsum()
plt.plot_date(dates, data, '-')
plt.title("Date Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=90)
plt.show()
```

## **4. Annotations**

### Annotations

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.annotate('Highest Point',
             xy=(iris_df["sepal length (cm)"].idxmax(),
                 iris_df["sepal length (cm)"].max()),
                 xytext=(100, 7),
                 arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

### Text

```python
plt.plot(iris_df["sepal length (cm)"])
plt.title("Sepal Length of Iris")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.text(50, 7.5, "Important Note", style='italic', bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
plt.show()
```
